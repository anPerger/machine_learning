```python
from sklearn import tree
import pandas as pd
import os
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("data/exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(30)
```

```python
# dropping candidates
no_candidates_df = df[~df["koi_disposition"].str.contains("CANDIDATE")]
no_candidates_df.head(30)  
```

```python
# train_test split on all features
X = no_candidates_df.drop(columns=["koi_disposition"])
y = no_candidates_df["koi_disposition"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# descision tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# feature importnce in descending order
sorted(zip(clf.feature_importances_, X), reverse=True)
```

```python
# random forest
rf = RandomForestClassifier(n_estimators=700)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# we can see a more even distribution of feature importance
sorted(zip(rf.feature_importances_, X), reverse=True)
```

```python
# limiting our features to those with a feature importance score of .01 or greater
over_1 = no_candidates_df[['koi_fpflag_co', 'koi_steff_err1', 'koi_prad', 'koi_steff_err2', 'koi_fpflag_ss', 'koi_fpflag_nt',
                           'koi_prad_err2', 'koi_duration_err2', 'koi_prad_err1', 'koi_duration_err1', 'koi_fpflag_ec', 
                           'koi_insol_err1', 'koi_time0bk_err2', 'koi_period', 'koi_model_snr', 'koi_insol_err1', 
                           'koi_time0bk_err2', 'koi_period', 'koi_srad_err1', 'koi_insol', 'koi_teq', 'koi_insol_err2',
                           'koi_impact', 'koi_depth', 'koi_period_err1', 'koi_slogg_err2', 'koi_period_err2']] 
over_1
```

```python
X = over_1
y = no_candidates_df["koi_disposition"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# with both descision and random forest models, there is little difference in accuracy score ...
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
sorted(zip(clf.feature_importances_, X), reverse=True)
```

```python
# ... when the features are limited to a score of .01 or greater
rf = RandomForestClassifier(n_estimators=700)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, X), reverse=True)
```
