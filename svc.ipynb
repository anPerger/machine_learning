```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(30)
```

```python
# drop the candidates
no_candidates_df = df[~df["koi_disposition"].str.contains("CANDIDATE")]
no_candidates_df.head(30)  
```

```python
# setting disposition binary for easier analysis later
no_candidates_df['koi_binary'] = no_candidates_df['koi_disposition'].apply(lambda x: 1 if x == 'CONFIRMED' else 0)
no_candidates_df
```

```python
# splitting x and y into train_test 
X = no_candidates_df.drop(columns=["koi_disposition"])
y = no_candidates_df["koi_disposition"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Scale your data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=['confirmed', 'false positivE']))
```
